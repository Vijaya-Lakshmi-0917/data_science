```python
heat_map = df[['QTY', 'VALUE']].corr()
f, ax = plt.subplots(figsize = (8,4))
sns.heatmap(heat_map, vmax=2.0, square = True, annot = True)
```

```python
Line_plot = df.groupby('QTY')['VALUE'].sum().reset_index()

plt.figure(figsize=(10, 6))
plt.plot(Line_plot['QTY'], Line_plot['VALUE'], marker='o', linestyle='-', color='blue')
plt.title('Line chart of quantity vs value')
plt.xlabel('QTY')
plt.ylabel('VALUE')
plt.grid(True)
plt.show()
```

```python
pd.crosstab(index = df['STORECODE'],
            columns = df['VALUE'],
            margins = True,
            normalize = 'columns')
```
